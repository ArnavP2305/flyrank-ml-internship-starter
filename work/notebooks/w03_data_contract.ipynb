### Five Feature Frame

| Feature | Available when? |
|---------|------------------|
| **gsc_impressions** | Available at the reporting date from Google Search Console, so it is known before making a prediction. |
| **gsc_avg_position** | Calculated from Search Console data already available at the reporting date. |
| **client_has_gsc** | Client configuration is known before prediction and does not depend on the outcome. |
| **client_has_ga4** | The client's GA4 integration status is known before prediction. |
| **gsc_data_available** | Indicates whether Search Console data exists for that row and is known at the decision moment. |

# The Trap(Feature Leakage)

```python
# Creating a small sample for the leakage demonstration
leak_df = feature_df.dropna().sample(n=10000, random_state=42)

# Deliberately creating a leaked feature
leak_df["leaked_clicks"] = leak_df["gsc_clicks"]

print("Leaked feature added successfully.")
print(leak_df[["gsc_clicks", "leaked_clicks"]].head())
```

```python
leak_df = leak_df.drop(columns=["leaked_clicks"])

print("Leaked feature removed.")
```

# Leakage Experiment

A new feature (leaked_clicks) was intentionally created by copying the target variable (gsc_clicks). This demonstrates feature leakage, where the model is given information that would not be available at prediction time. Such a feature would lead to unrealistically optimistic model performance, so it was removed before finalizing the feature set.
