## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

print("Total pages:", len(df))
print("Average CTR:", round(df["ctr"].mean(), 2))
print("Average impressions:", round(df["impressions_90d"].mean(), 2))
print("Pages with CTR < 1%:", (df["ctr"] < 1).sum())
print(
    "Pages with CTR < 1% and impressions > 1000:",
    ((df["ctr"] < 1) & (df["impressions_90d"] > 1000)).sum()
)
```

**The starter dataset contains 30,000 webpages, providing a sufficiently large sample for analysis. The average CTR is 0.51%, indicating that many pages receive relatively few clicks compared to their impressions. Additionally, 28,291 pages have a CTR below 1%, and 13,030 pages have both a CTR below 1% and more than 1,000 impressions. These observations suggest that there are many pages with visibility but relatively low engagement, making content refresh prioritization a meaningful research direction.**
